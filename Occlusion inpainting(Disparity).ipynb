```python
#Finding the mask for left disparity
import cv2
import imageio
import matplotlib.pyplot as plt 
img = cv2.imread('occludedLD.png')
mask = cv2.imread('mask.png')
res = cv2.bitwise_not(img,mask)
plt.imshow(res)
imageio.imwrite('maskLD1.png',res)
plt.show()
```

```python
#Finding the final mask for left disparity
import cv2
import imageio
import matplotlib.pyplot as plt
img = cv2.imread('occludedLD.png')
mask = cv2.imread('maskLD1.png')
res = cv2.bitwise_and(img,mask)
plt.imshow(res)
imageio.imwrite('maskLD2.png',res)
plt.show()
```

```python
#Applying inpainting (Method 1)
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('occludedLD.png')
mask = cv2.imread('maskLD2.png',0)

dst = cv2.inpaint(img,mask,3,cv2.INPAINT_NS)

plt.imshow(dst)
```

```python
#Applying inpainting (Method 2)
import numpy as np
import imageio
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('occludedLD.png')
mask = cv2.imread('maskLD2.png',0)

dst = cv2.inpaint(img,mask,3,cv2.INPAINT_TELEA)

plt.imshow(dst)
```

```python
imageio.imwrite('FinalLD.png',dst)
plt.show()
```
